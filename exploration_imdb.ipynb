```python
df_imdb_title_basics = pd.read_csv("title.basics.tsv", sep="\t")
```

```python
df_imdb_title_basics["titleType"].unique()
```

```python
df_imdb_title_basics = df_imdb_title_basics[df_imdb_title_basics["titleType"] == "movie"]
df_imdb_title_basics = df_imdb_title_basics[df_imdb_title_basics["isAdult"] == 0]
```

```python
df_imdb = pd.merge(df_imdb_title_basics,  # DataFrame de gauche
         df_imdb_ratings,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='tconst',  # Colonne clÃ© du DataFrame de gauche
         right_on='tconst')  # Colonne clÃ© du DataFrame de droite
```

```python
df_imdb_name_basics = pd.read_csv("name.basics.tsv", sep="\t")
```

```python
df_imdb_name_basics
```

```python
df_imdb_name = pd.merge(df_imdb_principals_actors_directors,  # DataFrame de gauche
         df_imdb_name_basics,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='nconst',  # Colonne clÃ© du DataFrame de gauche
         right_on='nconst')  # Colonne clÃ© du DataFrame de droite
```

```python
df_imdb_name
```

```python
df_imdb_final_ok = pd.merge(df_imdb,  # DataFrame de gauche
         df_imdb_name,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='tconst',  # Colonne clÃ© du DataFrame de gauche
         right_on='tconst')  # Colonne clÃ© du DataFrame de droite
```

```python
df_imdb_final_ok
```

```python
df_imdb_final = df_imdb_final_ok.drop(["birthYear","deathYear"], axis=1)
df_imdb_final = df_imdb_final_ok.drop(["titleType","isAdult", "endYear", "job", "knownForTitles"], axis=1)

```

```python
df_imdb_final
```

```python
df_imdb_final = df_imdb_final_ok.groupby(['tconst', 'primaryTitle', 'startYear', 'runtimeMinutes', 'genres','originalTitle', 'averageRating', "numVotes"], as_index=False).agg({'nconst': list, 'category': list, 'characters': list, "primaryName":list, "primaryProfession":list})
df_imdb_final.head()
```
