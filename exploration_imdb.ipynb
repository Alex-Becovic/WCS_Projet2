```python
df_imdb_title_basics = df_imdb_title_basics[df_imdb_title_basics["titleType"] == "movie"]
df_imdb_title_basics = df_imdb_title_basics[df_imdb_title_basics["isAdult"] == 0]
```

```python
df_imdb_title_basics
```

```python
df_imdb = pd.merge(df_imdb_title_basics,  # DataFrame de gauche
         df_imdb_ratings,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='tconst',  # Colonne clé du DataFrame de gauche
         right_on='tconst')  # Colonne clé du DataFrame de droite
```

```python
df_imdb
```

```python
df_imdb_name_basics = pd.read_csv("name.basics.tsv", sep="\t")
```

```python
df_imdb_final_ok = pd.merge(df_imdb,  # DataFrame de gauche
         df_imdb_name,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='tconst',  # Colonne clé du DataFrame de gauche
         right_on='tconst')  # Colonne clé du DataFrame de droite
```
