```python
# Traitement des valeurs manquantes
df['genres'] = df['genres'].fillna('')
df['primaryName'] = df['primaryName'].fillna('')
df['category'] = df['category'].fillna('')
```

```python
# Création d'une colonne combinée pour le filtrage basé sur le contenu
df['features_communes'] = df['genres'] + ' ' + df['primaryName'] + ' ' + df['category']
```

```python
df['features_communes'].head()
```

```python
df_film_echantillon = df.groupby(['tconst', 'primaryTitle', 'startYear', 'genres','originalTitle', 'averageRating', "numVotes"], as_index=False).agg({'nconst': list, 'category': list, 'features_communes': list, "primaryProfession":list})
df_film_echantillon.head()
```

```python
df_film_echantillon.to_csv("df_film_echantillon.csv")
```

```python
# Vectorisation des caractéristiques combinées
# La vectorisation consiste à transformer des informations textuelles en représentations numériques
# Pour que l'algorithme puisse comprendre et comparer ces informations, on les transforme en vecteurs numériques.
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
feature_matrix = vectorizer.fit_transform(df['features_communes'])
# sert à créer une nouvelle colonne features_communes en combinant les informations des colonnes 'genres',
# 'primaryName' et 'category'. Cette combinaison est essentielle pour les systèmes de recommandation basés sur le contenu, 
# car elle permet de représenter chaque film par un ensemble unifié de caractéristiques textuelles.
```
