```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('nettoyage_des_donnees.csv')
df.head(5)
```

```python
df_v1 = df.drop(['primaryName','category','overview','primaryTitle','production_companies_name','backdrop_path','primaryProfession','Unnamed: 0'],axis=1)
```

```python
df_v1.head(5)
```

```python
df_enc = df_v1["genres"].str.get_dummies(sep=",")
df_movies = pd.concat([df_v1, df_enc], axis=1)
df_movies = df_movies.drop(['genres'],axis=1)
```

```python
df_movies = pd.get_dummies(df_movies, columns=['original_language'])
```

```python
df_movies
```

```python
df_movies.head()
```

```python
scaler = MinMaxScaler()
df_merde = scaler.fit_transform(df_movies[['numVotes','runtimeMinutes','averageRating','startYear']])
df_merde
```

```python
daf = pd.DataFrame(df_merde, columns=['num_Votes', 'runtime_Minutes','average_Rating','start_Year'])

daf
```

```python
df_m = pd.concat([daf, df_movies], axis=1)
df_m.head()
```

```python
df_m = df_m.drop(['runtimeMinutes','numVotes','averageRating','startYear'], axis=1)
```

```python
df_m = df_m.set_index("originalTitle")
```

```python
df_m.head(1)
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=10)
```

```python
neigh.fit(df_m)
```

```python
distance,index_reco = neigh.kneighbors([df_m.loc['Superman']])
reco_index = index_reco[0]
noms_films = df_m.iloc[reco_index]
df_reco = df_m.iloc[reco_index]
df_reco["distance"] = distance[0]
display(df_reco)
```
