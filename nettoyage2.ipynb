```python
import pandas as pd

df = pd.read_csv('df_imdb.csv')
df.head(3)
```

```python
df_image = pd.read_csv('tmdb_full.csv')
```

```python
df_image = df_image[["imdb_id","backdrop_path"]]
```

```python
df_image
```

```python
df_imdb = pd.merge(df,  # DataFrame de gauche
         df_image,  # DataFrame de droite
         how="left",  # Type de jointure : ici, une jointure gauche (left join)
         left_on='imdb_id',  # Colonne clé du DataFrame de gauche
         right_on='imdb_id')  # Colonne clé du DataFrame de droite
```

```python
df_imdb
```

```python
# Convertion de la colonne 'startYear' en type numérique
df_imdb['startYear'] = pd.to_numeric(df['startYear'], errors='coerce')
```

```python
# Suppression les lignes avec des valeurs manquantes dans startYear
df_imdb = df_imdb.dropna(subset=['startYear'])
```

```python
# Convertir startYear en entier
df_imdb['startYear'] = df_imdb['startYear'].astype(int)
```

```python
# Filtrer les lignes où startYear est supérieur ou égal à 1950
df_imdb = df_imdb[df_imdb['startYear'] >= 1950]
df_imdb
```

```python
# Convertion de 'RuntimeWarnings' en type numérique
df_imdb['runtimeMinutes'] = pd.to_numeric(df_imdb['runtimeMinutes'], errors='coerce')
```

```python
# Suppression les lignes avec des valeurs manquantes dans startYear
df_imdb = df_imdb.dropna(subset=['runtimeMinutes'])
```

```python
# Convertir runtimeMinutes en entier
df_imdb['runtimeMinutes'] = df_imdb['runtimeMinutes'].astype(int)
```

```python
# Filtrer les lignes où startYear est supérieur ou égal à 1950
df_imdb = df_imdb[(df_imdb['runtimeMinutes'] >= 60) & (df_imdb['runtimeMinutes'] <= 240)]
df_imdb
```

```python
df_imdb = df_imdb[df_imdb['averageRating']>5]
df_imdb
```

```python
df_imdb = df_imdb.drop(['tconst','nconst','imdb_id'], axis=1)
```

```python
df_imdb = df_imdb[df_imdb['numVotes']> 100]
```
