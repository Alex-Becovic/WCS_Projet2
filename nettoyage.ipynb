```python
import pandas as pd

df = pd.read_csv('df_imdb.csv')
df.head(3)
```

```python
# Convertion de la colonne 'startYear' en type numérique
df['startYear'] = pd.to_numeric(df['startYear'], errors='coerce')
```

```python
# Suppression les lignes avec des valeurs manquantes dans startYear
df = df.dropna(subset=['startYear'])
```

```python
# Convertir startYear en entier
df['startYear'] = df['startYear'].astype(int)
```

```python
# Filtrer les lignes où startYear est supérieur ou égal à 1950
df_filtré = df[df['startYear'] >= 1950]
df_filtré
```

```python
# Convertion de 'RuntimeWarnings' en type numérique
df_filtré['runtimeMinutes'] = pd.to_numeric(df_filtré['runtimeMinutes'], errors='coerce')
```

```python
# Suppression les lignes avec des valeurs manquantes dans startYear
df_filtré = df_filtré.dropna(subset=['runtimeMinutes'])
```

```python
# Convertir runtimeMinutes en entier
df_filtré['runtimeMinutes'] = df_filtré['runtimeMinutes'].astype(int)
```

```python
# Filtrer les lignes où startYear est supérieur ou égal à 1950
df_filtré2 = df_filtré[(df_filtré['runtimeMinutes'] >= 60) & (df_filtré['runtimeMinutes'] <= 240)]
df_filtré2
```
